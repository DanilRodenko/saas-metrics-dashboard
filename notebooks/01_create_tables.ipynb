```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('postgresql+psycopg2://postgres:3963@localhost:5432/saas_metrics')
```

```python
customers = pd.read_csv("../data/raw/customers.csv")
subscription_events = pd.read_csv("../data/raw/subscription_events.csv")
```

```python
plans = pd.DataFrame({"plan_name": ["Starter", "Growth", "Enterprise"], "price": [19, 59, 249]})

customer_profile = customers[["customer_id", "company_name", "country", "acquisition_channel", "company_size_segment"]].rename(
    columns={"customer_id": "id"}
)



payment_profile = customers[["customer_id", "initial_plan", "payment_method", "is_auto_renew"]].rename(
    columns={"initial_plan": "current_plan"}
)

plans.to_sql('plans', engine, if_exists='append', index=False)
customer_profile.to_sql('customer_profile', engine, if_exists='append', index=False)
payment_profile.to_sql('payment_profile', engine, if_exists='append', index=False)
```

```python
transaction = subscription_events[["event_id", "customer_id", "plan", "event_type", "previous_plan","event_date", "is_auto_renew"]].rename(
    columns={"event_id": "transaction_id", "plan": "plan_name", "event_date": "transaction_date"}
)

transaction.to_sql('transaction', engine, if_exists='append', index=False)
```
